```python
import pandas as pd
import random
from datetime import datetime, timedelta

# 1. DimDate (2023-2024)
dates = []
start = datetime(2023,1,1)
for i in range(730):
    d = start + timedelta(days=i)
    dates.append({
        'DateKey': int(d.strftime('%Y%m%d')),
        'Date': d.strftime('%Y-%m-%d'),
        'Year': d.year,
        'Month': d.month,
        'MonthName': d.strftime('%B'),
        'Quarter': f'Q{(d.month-1)//3+1}',
        'Weekday': d.isoweekday(),
        'IsWeekend': 'Yes' if d.weekday() >= 5 else 'No'
    })
pd.DataFrame(dates).to_csv('DimDate.csv', index=False)

# 2. DimProduct (10 rows)
products = [
    ('BRK-101','Brake Rotor','Brakes',25.50,49.99),
    ('PAD-202','Brake Pad','Brakes',12.30,29.99),
    ('ENG-303','Engine Piston','Engine',45.00,89.99),
    ('ALT-404','Alternator','Electrical',65.20,129.99),
    ('BAT-505','Battery','Electrical',80.00,159.99),
    ('TRN-606','Transmission Gear','Transmission',120.50,249.99),
    ('SUS-707','Shock Absorber','Suspension',35.40,79.99),
    ('AXL-808','Axle','Chassis',95.60,189.99),
    ('RAD-909','Radiator','Cooling',110.30,219.99),
    ('FAN-010','Cooling Fan','Cooling',28.75,59.99)
]
prod_df = pd.DataFrame([(f'P{i+1:02d}', *p) for i,p in enumerate(products)], 
                       columns=['ProductKey','ProductCode','ProductName','ProductCategory','StandardCost','SellingPrice'])
prod_df.to_csv('DimProduct.csv', index=False)

# 3. DimMachine (6 rows)
machines = [
    ('Pressing-01','Line A','PressCo',3,500),
    ('Pressing-02','Line A','PressCo',2,500),
    ('CNC-Mill-01','Line B','CNCMaster',4,300),
    ('CNC-Mill-02','Line B','CNCMaster',1,300),
    ('Assembly-01','Line C','AutoAssy',2,400),
    ('Assembly-02','Line C','AutoAssy',0.5,400)
]
mac_df = pd.DataFrame([(f'M{i+1:02d}', *m) for i,m in enumerate(machines)],
                      columns=['MachineKey','MachineName','Line','Manufacturer','Age_Years','Capacity_PerHour'])
mac_df.to_csv('DimMachine.csv', index=False)

# 4. DimShift
shifts = [('S1','Morning','06:00','14:00','Regular'),
          ('S2','Evening','14:00','22:00','Regular'),
          ('S3','Night','22:00','06:00','Overtime')]
pd.DataFrame(shifts, columns=['ShiftKey','ShiftName','StartTime','EndTime','ShiftType']).to_csv('DimShift.csv', index=False)

# 5. DimEmployee (20 rows)
emps = [(f'E{i+1:03d}', f'Employee{i+1}', 'Operator' if i<15 else ('Supervisor' if i<18 else 'Quality Inspector'), 
         random.randint(1,15), round(random.uniform(15,35),2)) for i in range(20)]
pd.DataFrame(emps, columns=['EmployeeKey','EmployeeName','Role','Experience_Years','HourlyRate']).to_csv('DimEmployee.csv', index=False)

# 6. FactProduction (500 rows)
date_keys = pd.read_csv('DimDate.csv')['DateKey'].tolist()
prod_keys = pd.read_csv('DimProduct.csv')['ProductKey'].tolist()
mac_keys = pd.read_csv('DimMachine.csv')['MachineKey'].tolist()
shift_keys = pd.read_csv('DimShift.csv')['ShiftKey'].tolist()
emp_keys = pd.read_csv('DimEmployee.csv')['EmployeeKey'].tolist()

facts = []
for i in range(1,501):
    facts.append({
        'ProductionID': i,
        'DateKey': random.choice(date_keys),
        'ProductKey': random.choice(prod_keys),
        'MachineKey': random.choice(mac_keys),
        'ShiftKey': random.choice(shift_keys),
        'EmployeeKey': random.choice(emp_keys),
        'QuantityProduced': random.randint(200,500),
        'DefectCount': random.randint(2,20),
        'CycleTime_Min': round(random.uniform(1.5,5.0),1),
        'SetupTime_Min': random.randint(8,30),
        'Downtime_Min': random.randint(0,45)
    })
pd.DataFrame(facts).to_csv('FactProduction.csv', index=False)

print("All CSV files generated successfully!")
```
